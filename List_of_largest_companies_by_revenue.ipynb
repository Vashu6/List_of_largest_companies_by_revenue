```python
import requests
from bs4 import BeautifulSoup


# URL of the Zomato page you want to scrape
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'

# Send a GET request to the URL
response = requests.get(url)
```

```python
import pandas as pd
```

```python
page = requests.get(url)
```

```python
response
```

```python
soup1 = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup1.prettify())
```

```python
Title=soup1.find("table")
```

```python
Title
```

```python
title=Title.find_all("th")
```

```python
Main_Title=[T.text.strip() for T in title][1:9]
print(Main_Title)
```

```python
df=pd.DataFrame(columns=Main_Title,)
```

```python
df
```

```python
column_data=Title.find_all("tr")
```

```python
for row in column_data[2:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
#     print(individual_row_data)
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df['Rank'] = df.index + 1
```

```python
df
```

```python
df.to_csv(r'C:\Users\vikas\Desktop\Output\Companies.csv', index = False)
```
